```python
import pandas as pd
import matplotlib.pyplot as plt

filename = '/content/drive/MyDrive/run/runData.csv'
df = pd.read_csv(filename)

df
```

```python
# '날짜' 열을 datetime 형식으로 변환
df['날짜'] = pd.to_datetime(df['날짜'])

# '월' 열 추가
df['월'] = df['날짜'].dt.month

# 월별 달리기 거리 합산
monthly_distance = df.groupby('월')['거리'].sum()

# 월별 달리기 거리를 막대그래프로 표현
monthly_distance.plot(kind='bar', color='skyblue')
plt.xlabel('month')
plt.ylabel('distance (km)')
plt.title('monthly run distance')
plt.xticks(rotation=0)  # x축 레이블 회전

# 각 막대 안에 값을 표시
for i, v in enumerate(monthly_distance):
    plt.text(i, v + 0.1, str(round(v, 2)), ha='center', va='bottom', fontsize=8)

plt.show()
```

```python
# '날짜' 열을 datetime 형식으로 변환
df['날짜'] = pd.to_datetime(df['날짜'])

# '월' 열 추가
df['월'] = df['날짜'].dt.month

# 9월, 10월, 11월 데이터 추출
df_sep_oct_nov = df[(df['월'] >= 9) & (df['월'] <= 11)]

# 9월, 10월, 11월 중에서 내가 달린 날의 수 계산
days_run_sep_oct_nov = df_sep_oct_nov['날짜'].nunique()

# 9월, 10월, 11월 중에서 내가 달린 날의 수를 원 그래프로 표현
plt.pie([days_run_sep_oct_nov, 91 - days_run_sep_oct_nov], labels=['Days Run', 'Days Not Run'], autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Running Days in September, October, November')
plt.show()
```

```python
import matplotlib.pyplot as plt

# '날짜' 열을 datetime 형식으로 변환
df['날짜'] = pd.to_datetime(df['날짜'])

# '월' 열 추가
df['월'] = df['날짜'].dt.month

# 9월, 10월, 11월 데이터 추출
df_sep = df[df['월'] == 9]
df_oct = df[df['월'] == 10]
df_nov = df[df['월'] == 11]

# 각 월별 일별 달린 거리 계산
daily_distance_sep = df_sep.groupby(df_sep['날짜'].dt.day)['거리'].sum()
daily_distance_oct = df_oct.groupby(df_oct['날짜'].dt.day)['거리'].sum()
daily_distance_nov = df_nov.groupby(df_nov['날짜'].dt.day)['거리'].sum()

# 선 그래프로 표현
plt.plot(daily_distance_sep.index, daily_distance_sep, label='September', marker='o', linestyle='-', color='skyblue', alpha=0.7)
plt.plot(daily_distance_oct.index, daily_distance_oct, label='October', marker='o', linestyle='-', color='orange', alpha=0.7)
plt.plot(daily_distance_nov.index, daily_distance_nov, label='November', marker='o', linestyle='-', color='green', alpha=0.7)

plt.xlabel('Day of Month')
plt.ylabel('Distance (km)')
plt.title('Daily Running Distance from September to November')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)  # 그리드 추가
plt.show()
```
